```python
import csv
import random

# Function to generate a random Indian phone number
def generate_phone_number():
    return f"+91{random.randint(6000000000, 9999999999)}"

# Generate 1000 phone numbers
phone_numbers = [generate_phone_number() for _ in range(1000)]

# Save to a CSV file
with open("indian_phone_numbers.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Phone Number"])  # Column header
    for number in phone_numbers:
        writer.writerow([number])

print("CSV file 'indian_phone_numbers.csv' created successfully.")
```

```python
import csv

# Define the data
phone_numbers = [
    "1234567890",
    "9876543210",
    "1122334455",
    "5566778899",
    "9988776655"
]

# Define the output CSV file name
output_file = "sample_campaign.csv"

# Write the data to the CSV file
with open(output_file, mode="w", newline="") as file:
    writer = csv.writer(file)
    
    # Write the header
    writer.writerow(["id", "column1", "column3"])
    
    # Write the rows
    for idx, phone_number in enumerate(phone_numbers, start=1):
        writer.writerow([idx, phone_number, 1])

print(f"CSV file '{output_file}' created successfully!")
```
